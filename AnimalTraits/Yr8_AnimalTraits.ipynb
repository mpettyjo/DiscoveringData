#Linear Regression

Linear regression is a technique that models the relationship between a dependent variable and an independent variable by fitting a linear equation to observed data (known as the **line of best fit**. It not only describes the strength and direction of the relationship (like correlation) but also quantifies it with an equation, allowing for *prediction* of new data points.

Linear regression is an area of misconceptions for many learners of science and data science. It is important to understand the difference between linear regression and **correlation analysis**. Calculating a *correlation coefficient* measures how closely the changes in one variable are associated with changes in the other. It does not indicate that predictions can be made.

We sometimes see students reporting the equation of the line of best fit in situations where a predictive model would not be appropriate (eg see https://m.xkcd.com/2893/)




```python

# Ensure the selected variables exist and have enough data points
if filtered_df.empty:
    print("No data available for the selected variables.")
elif filtered_df.shape[0] < 2:
    print("Not enough data points to perform regression.")
else:
    # Perform linear regression using scipy
    slope, intercept, r_value, p_value, std_err = stats.linregress(
        filtered_df[x_var], filtered_df[y_var]
    )

    # Equation of the line and R² value
    line_eq = f'y = {slope:.2f}x + {intercept:.2f}'
    r_squared = f'R² = {r_value**2:.2f}'

    # Plot using sns.lmplot
    g = sns.lmplot(x=x_var, y=y_var, data=filtered_df, height=6, aspect=1.5, ci=None)

    # Add labels and title
    plt.xlabel(x_var.replace('_', ' ').title())
    plt.ylabel(y_var.replace('_', ' ').title())
    plt.title(f'Linear Fit of {x_var.replace("_", " ").title()} vs {y_var.replace("_", " ").title()}')

    # Add equation and R² value text on the plot
    plt.text(
        x=filtered_df[x_var].min(),
        y=filtered_df[y_var].max(),
        s=f'{line_eq}\n{r_squared}',
        fontsize=12,
        ha='left',
        va='top',
        bbox=dict(boxstyle='round,pad=0.5', edgecolor='black', facecolor='white')
    )

    # Show the plot
    plt.show()
```

Our interpretation here might suggest that different classes of animals have slightly different relationships.

Seaborn has a function called lmplot (linear model plot) that will quickly plot regression lines for data separated by category)

```python

# Ensure the selected variables exist and have enough data points
if filtered_df.empty:
    print("No data available for the selected variables.")
else:
    # Initialize the plot
    plt.figure(figsize=(10, 6))

    # Loop through each class and perform linear regression
    for class_name, group in filtered_df.groupby('class'):
        if group.shape[0] < 2:
            print(f"Not enough data points for class: {class_name}")
            continue

        # Plot the regression line for the current class
        sns.regplot(
            x=x_var, y=y_var, data=group, ci=None,
            scatter_kws={'alpha': 0.5}, label=class_name
        )

    # Add labels and title
    plt.xlabel(x_var.replace('_', ' ').title())
    plt.ylabel(y_var.replace('_', ' ').title())
    plt.title(f'Linear Fit of {x_var.replace("_", " ").title()} vs {y_var.replace("_", " ").title()} by Class')

    # Add a legend
    plt.legend(title='Class')

    # Show the plot
    plt.show()
```

Now we have nice tight-looking lines for each of our clusters. We still need to go back to the statsmodel to generate the information about our regression models.

```python

# List to store regression results for each class
regression_results = []

# Ensure there is data to work with
if filtered_df.empty:
    print("No data available for the selected variables.")
else:
    # Loop over each unique class to calculate regression statistics
    classes = filtered_df['class'].unique()
    for c in classes:
        class_df = filtered_df[filtered_df['class'] == c]

        # Check if there are at least two valid data points
        if class_df.shape[0] < 2:
            regression_results.append([c, "Insufficient Data", "N/A"])
            continue

        # Perform linear regression
        slope, intercept, r_value, _, _ = stats.linregress(
            class_df[x_var], class_df[y_var]
        )

        # Create the equation and R² text
        equation = f'y = {slope:.2f}x + {intercept:.2f}'
        r_squared = r_value**2

        # Append results to the list
        regression_results.append([c, equation, f'{r_squared:.2f}'])

    # Convert the results into a DataFrame for display
    results_df = pd.DataFrame(regression_results, columns=['Class', 'Equation', 'R²'])

    # Display the DataFrame
    display(results_df)
```
